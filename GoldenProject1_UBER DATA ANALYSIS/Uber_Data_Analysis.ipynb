```python
#most popular starting point for uber drivers

starting_point = uber_df['START*'].dropna()
df = pd.DataFrame(starting_point.value_counts())
df.sort_values(["START*"],ascending=False)
df
```

```python
df = df.reset_index()
df = df.rename(columns = {'index':'starting_destination', 'START*':'Count'})
df.loc[df['Count'] == max(df['Count'])]
```

```python
#least popular starting point for uber drivers

df = df.reset_index()
df = df.rename(columns = {'index':'starting_destination', 'START*':'Count'})
df.loc[df['Count'] == min(df['Count'])]
```

```python
#most popular dropping point for uber drivers

starting_point = uber_df['STOP*'].dropna()
df = pd.DataFrame(starting_point.value_counts())
df.sort_values(["STOP*"],ascending=False)

df = df.reset_index()
df = df.rename(columns = {'index':'starting_destination', 'STOP*':'Count'})
df.loc[df['Count'] == max(df['Count'])]
```

```python
#least popular dropping point for uber drivers

starting_point = uber_df['STOP*'].dropna()
df = pd.DataFrame(starting_point.value_counts())
df.sort_values(["STOP*"],ascending=False)

df = df.reset_index()
df = df.rename(columns = {'index':'starting_destination', 'STOP*':'Count'})
df.loc[df['Count'] == min(df['Count'])]
```

```python
#listing most frequent routes taken by uber drivers

df = uber_df.dropna()
df = pd.DataFrame(df.groupby(['START*','STOP*']).size())
df
```

```python
df = df.rename(columns = {0:'Count'})
df = df.sort_values(['Count'], ascending=False)
df.loc[df['Count'] == max(df['Count'])]
```

```python
#proportion of trips that is business and what is the proportion of trips that is personal

df = uber_df.groupby(['CATEGORY*']).sum()
Business = df.iloc[0,0]/(df.iloc[0,0] + df.iloc[1,0])
Personal = df.iloc[1,0]/(df.iloc[0,0] + df.iloc[1,0])

print("Business:", Business)
print("Personal:", Personal)
```
